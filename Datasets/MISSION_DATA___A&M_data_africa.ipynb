```python
# Filtrer les pays d'Afrique où le PIB est manquant
data_africa_sanspib = data_africa[data_africa['GDP (USD)'].isnull()]

# Regrouper par continent et obtenir les années où le PIB est manquant
pays_sanspib = data_africa_sanspib.groupby("Continent", as_index=False)['Year'].apply(lambda x: x.tolist())

# Renommer la colonne contenant les années manquantes
pays_sanspib = pays_sanspib.rename(columns={'Year': 'liste_années'})

# Ajouter une colonne pour compter le nombre d'années par continent
pays_sanspib['total_années'] = pays_sanspib['liste_années'].apply(len)

# Afficher le résultat
print(pays_sanspib)
```

```python
#Création d'un dataset avec la liste des pays membres de la CDEAO
#Liste des pays en anglais
df_cdeao = { "Country": [ "Benin", "Burkina Faso", "Cape Verde", "Ivory Coast", "Gambia", "Ghana", "Guinea", "Guinea-Bissau", "Liberia", "Mali", "Niger", "Nigeria", "Senegal", "Sierra Leone", "Togo" ] }
df_cdeao["Country"]
```

```python
#liste des pats pour lesquels n a des données manquantes
list_manquant = data_africa_sanspib['Country'].unique().tolist()
list_manquant
```

```python
#Nous allons supprimer les pays pour lesquels il manques des données et qui ne font pas partie de la CDEAO
a_supprimer = []
for i in list_manquant:
  if i not in df_cdeao["Country"]:
    a_supprimer.append(i)
```

```python
a_supprimer
```

```python
#Supprimer toutes les lignes dans data_africa qui contiennent les données dans a_supprimer
# ~ intègre la notion de "A CHAQUE FOIS QUE"
dfclean = data_africa[~data_africa["Country"].isin(a_supprimer)]
dfclean.shape
```

```python
len(dfclean['Country'].unique())
```

```python
liste_finale = dfclean['Country'].unique().tolist()
liste_finale
```

```python
#création d'une fonction pour préciser s'il s'agit de la CDEAO ou non
def cdeao_membership (liste_pays, list_cdeao):
  resultat = []
  for i in liste_pays:
    if i in list_cdeao:
      resultat.append("OK")
    else:
      resultat.append("KO")
  return resultat
```

```python
#Création d'une colonne CDEAO_member dans le df communiqué par le client
dfclean['CDEAO_members'] = cdeao_membership(dfclean['Country'].tolist(), df_cdeao["Country"])
dfclean.head()
```

```python
#Pays de la CDEAO avec des valeurs nulles == >Objectif trouver les indicaterrs et les ajouter dans le dataframe

cdeao_manquants = dfclean[dfclean.isnull().any(axis=1) & (dfclean['CDEAO_members'] == "OK")]
cdeao_manquants
```
